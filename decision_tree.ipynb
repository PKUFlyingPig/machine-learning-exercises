```python
from sklearn import tree
import csv
import graphviz

X = []
Y = []
feature_names=['色泽','根部','敲声','纹理','脐部','触感','密度','含糖率']
classes_names=["好瓜","坏瓜"]
with open("watermelon.csv") as file:
    reader = csv.reader(file)
    for line in reader:
        X.append(list(map(float, line[:-1])))
        Y.append(int(line[-1]))

clf = tree.DecisionTreeClassifier()
clf.fit(X, Y)
dot_data = tree.export_graphviz(clf, out_file=None,
                               feature_names=feature_names,
                               class_names=classes_names,
                               filled=True,rounded=True,
                               special_characters=True)
graph = graphviz.Source(dot_data)

```

```python
from sklearn import tree
import matplotlib.pyplot as plt
import numpy as np
import graphviz
X=np.random.rand(100)*np.pi
noise=np.random.normal(size=100)/20
Y=np.sin(X)+noise
plt.plot(X,Y,'bo')
plt.show()
X=X.reshape(-1,1)
clf = tree.DecisionTreeRegressor(max_depth=10)
clf.fit(X, Y)
dot_data = tree.export_graphviz(clf, out_file=None)
graph = graphviz.Source(dot_data,)
test_x=np.arange(0,np.pi,0.0001).reshape(-1,1)
pred=clf.predict(test_x)
plt.plot(test_x,pred,'ro',ms=2)
plt.show()
```
